# Temporal Analysis

Graph #1: Number of earthqaukes annually

```python
# I found that there were a few years that were incorrectly read from Excel, and saw other values were tainted
# for those data point too, so I deleted those rows.

non_integer_rows = df[~df['Year'].astype(str).str.isdigit()]
non_integer_rows
```

```python
df = df.drop([6786, 16597])
df = df[:19905]
```

```python
# Looking at the number of earthqaykes per year. As we can see, the number has dramaticallly increased.

counts_year = df.groupby('Year').size().reset_index(name='Count')

plt.plot(counts_year['Year'], counts_year['Count'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.title('Number of Earthquakes by Year')
plt.show()
```

Graph #2: Number of earthqaukes per month

```python
counts_month = df.groupby('Month').size().reset_index(name='Count')
counts_month['months'] = ["January", "February", "March", "April", "May", "June", "July", "August", "September","October",
                      "November", "December"]


plt.plot(counts_month['months'], counts_month['Count'], marker='o', linestyle='-', color='b')
plt.xticks(rotation = 45)
plt.ylabel('Number of Earthquakes')
plt.title('Number of Earthquakes by Month')
plt.show()
```

# Average Largest Magnitude per Year


```python
avg_year = df.groupby('Year').aggregate({'Largest Magnitude':'mean'})

plt.plot(avg_year, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Average Largest Magnitude')
plt.title('Average Largest Magnitude by Year')
plt.show()
```

# Magnitude Heat Map

I have created a visual represention of the "hottest earthqauke zones" in Turkey spanning 1915 to 2023, where the largest magnitude is measured for the earthqaukes.

```python
map = go.Densitymapbox(lat=df['Latitude'], lon=df['Longitude'], z=df['Largest Magnitude'], radius=10, colorscale='YlOrRd',showscale=True)

layout = go.Layout(title='Earthquake Density Heatmap', mapbox=dict(style='stamen-terrain',center=dict(lon=35, lat=39),zoom=5))

fig = go.Figure(data=[map], layout=layout)

fig.show()
```
